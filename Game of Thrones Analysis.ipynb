# Analyzing the Book: Game of Thrones

## This text file 'gameofthrones' contains books 1-5 from the Game of Thrones Boxed Set, including 'A Game of Thrones,' 'A Clash of Kings,' 'A Storm of Swords,' and 'A Feast for Crows.'" Our objective in this analysis is to identify the top 10 main characters in the book series.

### First step, we read in the file

```python
with open('gameofthrones.txt', 'r', encoding='utf-8') as f:
    lines = f.readlines()
    print("File read in successfully.")
```

### Next step is counting the frequency of each word, line by line.

```python
d = {}

for l in lines:
    l = l.strip()
    if not l:
        continue 
    words = l.split()
    caps = [w.rstrip('.’,;?!:"') for w in words if w[0].isupper()]
    caps = [w.replace("\'s", '') for w in caps]
    for n in caps:
        if n in d: 
            d[n] += 1
        else:
            d[n] = 1

mains = sorted(d.items(), key=lambda x: x[1], reverse=True)
mains[:25]
```

```python
# These are the most common sentence-starting words that we will exclude:
excluded = ['I', "I'll",  'I’ll', 'In', 'It', 'On', 'We', 'As', 'The', 'A', 'He', 'His', 'An', 'She', 'But', 'And', 'Of', 'This', 'They', 'Mr', 'Mrs', 'Miss', "I'd", "I'm"]
```

```python
mains = [(word, count) for word, count in mains if word not in excluded]
mains[:25]
```

### Now we see more words we need to remove like: Lord, Lady, You, That, There, Ser, When, etc

```python
#Updating Excluded words

excluded.extend(['Lord', 'Lady', 'Ser', 'When', 'You', 'If', 'Her', 'That', 'There', 'No', 'Your', 'How'])
```

```python
mains = [(word, count) for word, count in mains if word not in excluded]
mains[:25]
```

```python
names = [x[0] for x in mains[:10]] 
counts = [x[1] for x in mains[:10]] 
top_characters = dict(zip(names, counts))
top_characters
```

## Using Seaborn for Visualization

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=names, y=counts, palette="cividis")
plt.xlabel('Names of Characters', fontsize=14)
plt.ylabel('Counts')
plt.title('Top 10 Most Mentioned Names in "The Game of Thrones"')
plt.show()
```
